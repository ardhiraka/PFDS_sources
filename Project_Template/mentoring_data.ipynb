## Data Preparation

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('googleplaystore.csv')

print ('Data read into a pandas dataframe!')
```

```python
df.head()
```

```python
df.shape
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(6)
```

```python
df.dropna(how ='any', inplace = True)
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(6)
```

## Highest and Lowest Apps

```python
apps_paid=df[df['Type']=='Paid']
```

```python
apps_paid.head()
```

```python
apps_paid.Price = apps_paid.Price.apply(lambda x: x.replace('$',''))
apps_paid['Price'] = apps_paid['Price'].apply(lambda x: float(x))
```

```python
apps_paid.Price.max()
```

```python
apps_paid[apps_paid['Price'] == 400]
```

Apps with highest price 

```python
apps_paid.sort_values('Price',ascending=False).head(10)
```

```python
apps_paid.sort_values('Price',ascending=True).head(10)
```
